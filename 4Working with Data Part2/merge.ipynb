```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
```

```python
dframe1 = DataFrame({'key':['A','B','C','D','E','F','G'], 'data_set1': np.arange(7)})
```

```python
dframe1
```

```python
dframe2 = DataFrame({'key':['A','B','C'], 'data_set2': [1,2,3]})
```

```python
dframe2
```

```python
dframe2['data_set2']
```

```python
df = pd.merge(dframe1,dframe2)
```

```python
df
```

```python
df1 = pd.merge(dframe1,dframe2,on='key')
```

```python
df1
```

```python
df2 = pd.merge(dframe1,dframe2,on='key', how='left')
```

```python
df2
```

```python
df3 = pd.merge(dframe1,dframe2,on='key', how='right')
```

```python
df3
```

```python
df4 = pd.merge(dframe1,dframe2,on='key', how='outer')
```

```python
df4
```

```python
df_left = DataFrame({'key1': ['SF','AUS','IND'],
                    'key2': ['one','two','three'],
                    'left_data': [200,300,400]})
```

```python
df_left
```

```python
df_right = DataFrame({'key1': ['SF','SF','SF','AUS','AUS','IND','IND','IND'],
                    'key2': ['one','one','one','two','two','three','three','three'],
                    'left_data': [150,200,330,350,345,400,415,498]})
```

```python
df_right
```

```python
pd.merge(df_right,df_left,on=['key1','key2'], how='outer')
```

```python
pd.merge(df_right,df_left,on=['key1','key2'], how='left')
```

```python
pd.merge(df_right,df_left,on=['key1','key2'], how='inner')
```

```python
pd.merge(df_right,df_left,on='key1')
```

```python
pd.merge(df_right,df_left,on='left_data')
```

```python
pd.merge(df_right,df_left,on='key1', suffixes=('_lefty','_righty'))
```
