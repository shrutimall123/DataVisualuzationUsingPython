```python
import yfinance as yf
import pandas as pd
import datetime
import time

tickers = ['CVX', 'XOM', 'BP']
start_date = datetime.datetime(2010, 1, 1)
end_date = datetime.datetime(2023, 1, 1)

prices = pd.DataFrame()

for ticker in tickers:
    data = yf.download(ticker, start=start_date, end=end_date)['Adj Close']
    prices[ticker] = data
    time.sleep(1)  # Add a 1-second delay between downloads

print(prices.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'rets' is your DataFrame with returns data
correlation_matrix = rets.corr()

# Create a new figure with a specific size
plt.figure(figsize=(10, 8))

# Create the heatmap
sns.heatmap(correlation_matrix, 
            annot=True,  # This will add the values in the squares
            cmap='viridis',  # You can keep 'viridis' or choose another colormap
            fmt='.2f',  # Format the numbers to show only two decimal places
            linewidths=0.5,  # Add lines between the squares
            cbar=True)  # Include a color bar

# Set the title
plt.title('Correlation Matrix of Returns', fontsize=16)

# Adjust the layout and display the plot
plt.tight_layout()
plt.show()
```
